### Gives true if a number is prime

```python
from math import sqrt

def is_prime(num):
    num_range = int(sqrt(num)) + 1
    divisors = [n for n in range(2, num_range) if num%n==0]
    return len(divisors) == 0
```

### Same as above but consumes less time

```python
from math import sqrt
def isprime(num):
    if num == 0 or num == 1 or not num % 2:
        if num == 2:
            return True
        return False
    for i in range(3, int(sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True
```
