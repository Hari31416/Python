```python
with open('albums_ed.json', 'r') as f:
    data = json.load(f)
```

```python
names = []
uris = []
for item in data['items']:
    name = item['name']
    uri = item['uri']
    uri = uri.split(':')[-1]
    names.append(name)
    uris.append(uri)

uri_set = set(uris)
names_set = set(names)
```

```python
df = pd.DataFrame(data = [names, uris]).transpose()
df.columns = ['Albums', 'URI']
df_final = df.drop_duplicates(subset=['Albums'])

directories = list(df['Albums'].unique())
```
