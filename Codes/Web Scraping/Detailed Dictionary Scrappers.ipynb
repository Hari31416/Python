```python
def get_meaning(word):
    """
    Returns the list of meanings
    Maximum number is 5
    If position = 2 does not work, try using 0.
    """
    
    link = f'https://www.merriam-webster.com/dictionary/{word}'
    code = requests.get(link)
    soup = bs4.BeautifulSoup(code.text, "lxml")
    
    #Creating the Meaning list
    i = 0
    meanings = []

    raw_meanings = soup.select('.sense')
    #First Method
    for article in raw_meanings:
        #Choosing only relevant meanings
        if article.text[1].isdecimal() or article.text[2] == ':':
            i += 1
            meaning = article.text.strip().split('\n')[2]
            
            try:
                meaning = meaning.split(':')[1].strip()
                meanings.append(meaning)
            except IndexError:
                break

            #Ignoring meanings when number in greater than 5
            if i > 4:
                break

    #Sometimes the above method does not work
    #The following method is for that case
    if len(meanings) == 0:
        for article in raw_meanings:
            #Choosing only relevant meanings
            if article.text[1].isdecimal() or article.text[2] == ':':
                i += 1
                meaning = article.text.strip().split('\n')[0]

                try:
                    meaning = meaning.split(':')[1].strip()
                    meanings.append(meaning)
                except IndexError:
                    break

                #Ignoring meanings when number in greater than 5
                if i > 4:
                    break
                    
    #sorting the list by number of words
    for meaning in meanings:
        if len(meaning.split(' '))<3:
            
            x = {}
            for meaning in meanings:
                i = len(meaning.split(' '))
                x[meaning] = i

            x = {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}

            return list(x.keys())
        else:
            return meanings
```

```python
def syns_and_ants():
    """
    Returns lists of synonyms and Antonyms as a tuple
    """
    
    # Synonyms and Antonyms
    synonyms =[]
    antonyms = []

    others = soup.select('.mw-list')
    syns = others[0].text.strip()
    ants = others[-1].text.strip()

    for syn in syns.split(','):
        synonyms.append(syn.strip())

    for ant in ants.split(','):
            antonyms.append(ant.strip())
    synonyms = ', '.join(synonyms)
    antonyms = ', '.join(antonyms)
    return synonyms, antonyms
```

```python
def choose_the_right_synonym():
    """
     Choose the Right Synonym (not available for all)
     Returns a string
    """
    
    #The grand list
    grand_list = []
    the_list = soup.select('.syn')[0].text.strip().split('.')
    for sent in the_list:
        if sent.split('\n'):
            the_list2 = sent.split('\n')
            for obj in the_list2:
                if obj.strip() != '':
                    grand_list.append(obj.strip())
        else:
            grand_list.extend(sent)


    # modifying the grand list to look pretty

    choose = ''
    for sent in grand_list:

        if grand_list.index(sent) == 0 or grand_list.index(sent)%2 ==0:
            choose +=sent.strip() +  '\n\n' 
        else:
            choose +=sent.strip() +  '\n'
    choose = choose.strip()
    return choose
```

# The Final Function

```python
def add_together(word_list):
    """
    A function aggregating all the above functions.
    """

    #Containers
    words = []
    meaning1 = []
    meaning2 = []
    meaning3 = []
    meaning4 = []
    meaning5 = []
    synonyms_list = []
    antonyms_list = []
    choose_right = []
    
    total_len = len(word_list)
    print(f'Total words given is: {total_len}.')
    print('Remaining:')
    
    for word in word_list:

        link = f'https://www.merriam-webster.com/dictionary/{word}'
        code = requests.get(link)
        soup = bs4.BeautifulSoup(code.text, "lxml")
        index = word_list.index(word)
        print(f"\t{total_len-index}")
        
        
        #Creating the Meaning list
        i = 0
        meanings = []

        raw_meanings = soup.select('.sense')

        for article in raw_meanings:
            #Choosing only relevant meanings
            if article.text[1].isdecimal():
                i += 1
                meaning = article.text.strip().split('\n')[2]
                try:
                    meaning = meaning.split(':')[1].strip()
                    meanings.append(meaning)
                except IndexError:
                    break

                #Ignoring meanings when number in greater than 5
                if i > 4:
                    break
                    
        if len(meanings) == 0:
            for article in raw_meanings:
                #Choosing only relevant meanings
                if article.text[1].isdecimal() or article.text[2] == ':':
                    i += 1
                    meaning = article.text.strip().split('\n')[0]

                    try:
                        meaning = meaning.split(':')[1].strip()
                        meanings.append(meaning)
                    except IndexError:
                        break

                    #Ignoring meanings when number in greater than 5
                    if i > 4:
                        break

        #sorting the list by number of words
        for meaning in meanings:
            if len(meaning.split(' '))<3:

                x = {}
                for meaning in meanings:
                    i = len(meaning.split(' '))
                    x[meaning] = i

                x = {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}

                meanings =  list(x.keys())
            else:
                pass

            
        # Synonyms and Antonyms
        try:
            synonyms =[]
            antonyms = []

            others = soup.select('.mw-list')
            syns = others[0].text.strip()
            ants = others[-1].text.strip()

            for syn in syns.split(','):
                synonyms.append(syn.strip())

            for ant in ants.split(','):
                    antonyms.append(ant.strip())
            synonyms = ', '.join(synonyms)
            antonyms = ', '.join(antonyms)
        except:
            synonyms = 'Oops! So Empty!'
            antonyms = 'Oops! So Empty!'

            
        # Choose the Right Synonym
        try:
            #The grand list
            grand_list = []
            the_list = soup.select('.syn')[0].text.strip().split('.')
            for sent in the_list:
                if sent.split('\n'):
                    the_list2 = sent.split('\n')
                    for obj in the_list2:
                        if obj.strip() != '':
                            grand_list.append(obj.strip())
                else:
                    grand_list.extend(sent)

            # modifying the grand list to look pretty
            choose = ''
            for sent in grand_list:

                if grand_list.index(sent) == 0 or grand_list.index(sent)%2 ==0:
                    choose +=sent.strip() +  '\n\n' 
                else:
                    choose +=sent.strip() +  '\n'
            choose = choose.strip()
        except IndexError:
            choose = 'Oops! So Empty!'

            
        #Creating the df
        #The Word
        words.append(word)

        #Meanings
        try:
            meaning1.append(meanings[0])
        except IndexError:
            meaning1.append('Oops! So Empty!')

        try:
            meaning2.append(meanings[1])
        except IndexError:
            meaning2.append('Oops! So Empty!')

        try:
            meaning3.append(meanings[2])
        except IndexError:
            meaning3.append('Oops! So Empty!')

        try:
            meaning4.append(meanings[3])
        except IndexError:
            meaning4.append('Oops! So Empty!')

        try:
            meaning5.append(meanings[4])
        except IndexError:
            meaning5.append('Oops! So Empty!')

        #Synonyms and Antonyms
        synonyms_list.append(synonyms)
        antonyms_list.append(antonyms)

        #Right Synonym
        choose_right.append(choose)

    #List of columns
    data = [words, meaning1, meaning2, meaning3, meaning4, meaning5, synonyms_list, antonyms_list, choose_right]
    name_rows = 'Word, First, Second, Third, Fourth, Fifth, Synonyms, Antonyms, Choose the Right Synonym'.split(', ')
    df = pd.DataFrame(data=data, index=name_rows).transpose()
    return df
```
