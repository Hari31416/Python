```python
import requests
import bs4
```

```python
src = requests.get('https://en.wikipedia.org/wiki/Brandon_Sanderson_bibliography')
```

```python
soup = bs4.BeautifulSoup(src.text, "lxml")
```

```python
#soup
```

```python
the_list = soup.select(".wikitable sortable jquery-tablesorter, tr>td>i")
```

```python
books = []
for text in set(the_list):
    books.append(text.text)
```

```python
book_Brandon = list(set(books))
```

```python
sorted(book_Brandon)
```

```python
len(book_Brandon)
```

```python
src = requests.get("https://en.wikipedia.org/wiki/Isaac_Asimov_bibliography_(alphabetical)")
soup = bs4.BeautifulSoup(src.text, "lxml")
```

```python
the_list = soup.select("ul li")
```

```python
book_asimov = []
for item in the_list:
    book_asimov.append(item.text)
```

```python
book_asimov = set(book_asimov)
```

```python
sorted(book_asimov)
```

```python
len(book_asimov)
```
