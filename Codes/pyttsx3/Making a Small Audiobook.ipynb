```python
import pyttsx3
import re
import os
import shutil
engine = pyttsx3.init()
```

```python
with open("PandP.txt", 'r', encoding = "utf-8") as f:
    raw_text = f.read()
```

```python
pattern = r"\n(Chapter)\s(.*?)\n"
chapters = {}
find_iter = re.finditer(pattern, raw_text)
current_iter = next(find_iter)
start_current_chapter = current_iter.span()[0]
while True:
    try:
        chapter = current_iter.group(2)
        chapter = f"Chapter {chapter}"
        print(f"Extracting {chapter}")
        current_iter = next(find_iter)
        end_current_chpater = current_iter.span()[0]
        current_chapter_text = raw_text[start_current_chapter:end_current_chpater].strip()
        chapters[chapter] = current_chapter_text
        start_current_chapter = end_current_chpater
    except Exception as e:
        current_chapter_text = raw_text[start_current_chapter:]
        chapters[chapter] = current_chapter_text
        print("No more chapters")
        break
```

```python
engine.setProperty('rate', 150)
voices = engine.getProperty('voices')
engine.setProperty("voice", voices[2].id)
os.makedirs("Pride And Prejudice", exist_ok=True)
for key, value in chapters.items():
    print(f"{key}")
    engine.save_to_file(value, f"{key}.mp3")
    engine.runAndWait()
    current_path = os.path.join(os.getcwd(), f"{key}.mp3")
    shutil.move(current_path, "Pride And Prejudice")
```
