```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from ipywidgets import interactive
from IPython.display import display
from matplotlib import animation
```

```python
def vector_field(w,t,p):
    """
    Returns the value of the vector field at the point (x, y, t)
    """
    y, y1 = w
    k = p
    return np.array([
        y1,
        -k**2*y
    ])
```

```python
from scipy.integrate import odeint
k = 0.5
t = np.linspace(-10, 10, 100)
y0 = [0.1, 0]
w = odeint(vector_field, y0, t, args=(k,))

w1 = w[:,0]
w2 = w[:,1]
plt.plot(t,w1,label='y')
plt.plot(t,w2,label='y1')
plt.legend()
```

### Solving
$$y'' + \sin{x}y' + 2y = 0,$$
Let $y_1 = y$
$$y_1' = y_2$$
$$y_2^{'} = -2y_1 - \sin{x}y_2$$

```python
def vector_field(w,t):
    """
    Returns the value of the vector field at the point (x, y, t)
    """
    y1, y2 = w
    return np.array([
        y2,
        -2*y1-y2*np.sin(t)
    ])

def solvr(Y, t):
    return [Y[1], -2 * Y[0]-Y[1]]
```

### Solving
$$xy''+(c-x)y''-ay=0,$$
Again, let $y_1 = y$
$$y_1' = y_2$$
$$y_2^{'} = ay/x -(c-x)y_1'$$

```python
def vector_field(w,t, p):
    """
    Returns the value of the vector field at the point (x, y, t)
    """
    y1, y2 = w
    a, c = p
    return np.array([
        y2,
        a*y1/t- (c-t)*y2,
    ])
```
