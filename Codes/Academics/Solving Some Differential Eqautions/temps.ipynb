```python
import pandas as pd
```

```python
tables = pd.read_html('file:///C:/Users/harik/Desktop/Python/Codes/Academics/Solving%20Some%20Differential%20Eqautions/temp.html')
```

```python
df = tables[0]
```

```python
df_imp = df[['Unnamed: 0', 'Aspect ratio']]
```

```python
df_imp['Aspect ratio'][0].split(':')
```

```python
def get_aspect_ratio(row):
    nums= row['Aspect ratio'].split(':')
    return float(nums[0])/float(nums[1])
```

```python
df_imp['Aspect ratio Calc'] = df_imp.apply(get_aspect_ratio, axis=1)
```

```python
df_imp
```

```python
candidates = df_imp[df_imp['Aspect ratio Calc']<1.35]
```

```python
candidates.to_csv('candidates.csv')
```
