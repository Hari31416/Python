```python
# %matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2, 100)
y = np.sin(2 * np.pi * x)
z = np.cos(2 * np.pi * x)
fig = plt.figure()
plt.plot(x, y)
plt.scatter(x, z)
```

```python
def potential(x, n = 2, k=1e-10, l=1, mu=1e12):
    return k*x**n + l**2/(2*mu*x**2)
```

```python
def potential_new(x, n=1):
    return -x**n+x**(-2)
```

```python
%matplotlib widget
plt.figure(figsize=(8,4))         
t = np.linspace(0.001, 10, 200)
plt.plot(t, potential_new(t, n=-3))
plt.plot(t, np.zeros(len(t)))
# plt.xlim(5, 10)
# plt.ylim(-1, 1)
```
