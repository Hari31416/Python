```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2, 100)
y = np.sin(2 * np.pi * x)
z = np.cos(2 * np.pi * x)
fig = plt.figure()
plt.plot(x, y)
plt.scatter(x, z)
```
