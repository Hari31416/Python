```python
the_dir = 'C:\\Users\\harik\\Desktop\\Option\\'
i = 0
for image in images:
    image1 = Image.open(the_dir+image)
    
    x = 100
    y=100
    w = 750
    h = 700
    info = image1.crop((x,y,x+w, y+h))
#     info.save(fp=the_dir+'Info\\'+f"info_{i}.png", format='PNG')
    
    x = 1020
    y=850
    w = 600
    h = 350
    panl = image1.crop((x,y,x+w, y+h))
#     panl.save(fp=the_dir+'PanL\\'+f"panl_{i}.png", format='PNG')
    
    blank = Image.new(size=(1400,650), mode = 'RGB', color=(255,255,255))
    blank.paste(info, (0,0))
    blank.paste(panl ,(800,150))
    blank.save(fp=the_dir+f"img_{i}.png", format='PNG')
    i+=1
```
