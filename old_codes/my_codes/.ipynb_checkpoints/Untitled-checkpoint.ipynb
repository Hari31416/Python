```python
from random import choice
class RandomWalk():
	"""a simple model of random walk problem"""
	def __init__(self, steps):
		self.steps = steps
		"""setting initial positions"""
		self.x_o = [0]
		self.y_o = [0]
	def multiple_step(self):
		"""running single_step a number of times"""
		for n in range(1, self.steps):
			x_value = choice([1, -1])
			y_value = choice([1, -1])
			self.x_o.append(self.x_o[-1] + x_value)
			self.y_o.append(self.y_o[-1] + y_value)
rw = RandomWalk(100000)
rw.multiple_step()
import matplotlib.pyplot as plt
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(rw.x_o, rw.y_o)
plt.show()
```

```python
import random
def prime(n):
    c = []
    d = []
    for m in range (1,1000):
        a = random.randint(1, n)
        b = random.randint(1, n)
        if a%b == 0:
            c.append(m)
        if a%b != 0:
            d.append(m)
    from plotly.graph_objs import Bar, Layout
    from plotly import offline
    x_values = ['yes', 'no']
    y_values = [len(c), len(d)]
    data = [Bar(x = x_values, y = y_values)]
    x_axis_config = {'title': 'Result'}
    y_axis_config = {'title': 'Frequency of Result'}
    my_layout = Layout(title='Results of divisibility of two random number', 
        xaxis=x_axis_config, yaxis=y_axis_config)
    offline.plot({'data': data, 'layout': my_layout}, filename='d6.html')
prime(10)
```
