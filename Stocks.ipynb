```python
def fibo_retrace(maximum, minimum):
    levels = [0.14, 0.25, 0.38, 0.5, 0.62, 0.75]
    diff = maximum-minimum
    for i in range(len(levels)):
        print(maximum - diff*levels[i])
```

```python
import pandas as pd
import numpy as np

#Containers
start_dates = []
end_dates = []
profits = []
week_profits = []
day_profits = []
days = []
weeks = []
start_prices = []
last_prices = []

# The Calculator
def calc_indi():
    
    #Variables
    while True:
        try:
            start = input('The Starting Date: ')
            start_date = pd.to_datetime(start)
            break
        except ValueError:
            print('Nope!')
            continue

    while True:
        try:
            start_price = float(input('The Start Price: '))
            break
        except TypeError and ValueError:
            print('Nope!')
            continue
           
    while True:
        try:
            end = input('The Ending Date: ')
            end_date = pd.to_datetime(end)
            break
        except ValueError:
            print('Nope!')
            continue
    
    while True:
        try:
            last_price = float(input('The Final Price: '))
            break
        except TypeError and ValueError:
            print('Nope!')
            continue
    
    # The Gain
    profit = (-start_price+last_price)*100/start_price
    
    #Number of Weeks
    delta = end_date - start_date
    num_of_days = delta.days + 1
    num_of_weeks = int(np.ceil(delta.days/7))
    
    #Profit per day and week
    per_day = profit/num_of_days
    per_week = profit/num_of_weeks
    
    #Filling the lists
    start_dates.append(start_date.date())
    end_dates.append(end_date.date())
    profits.append(profit)
    week_profits.append(per_week)
    day_profits.append(per_day)
    days.append(num_of_days)
    weeks.append(num_of_weeks)
    start_prices.append(start_price)
    last_prices.append(last_price)

# Continue Function
def conti():
    start = input('Continue? ')
    return start[0].lower() == 'y'

while True:
    calc_indi()
    if not conti():
        break

#Creating the DataFrame
df = pd.DataFrame(data = [start_dates, end_dates, start_prices, last_prices, days, weeks, week_profits, day_profits, profits], index='Start Date, End Date, Bought Price, Sold Price, Days, Approx. Weeks, Weekly, Daily, Total Profit'.split(', ')).transpose()
df
```

```python
import pandas as pd
import numpy as np

def calc_all(start_dates_, end_dates_, start_prices, last_prices):
    """
    Lists Required: start_dates_, end_dates_, start_prices, last_prices
    """
    
    profits = []
    week_profits = []
    day_profits = []
    days = []
    weeks = []
    start_dates = []
    end_dates = []

    for i in range(len(start_dates_)):
        start_date = pd.to_datetime(start_dates_[i])
        start_price = float(start_prices[i])
        end_date = pd.to_datetime(end_dates_[i])
        last_price = float(last_prices[i])

        # The Gain
        profit = (-start_price+last_price)*100/start_price

        #Number of Weeks
        delta = end_date - start_date
        num_of_days = delta.days + 1
        num_of_weeks = int(np.ceil(delta.days/7))

        #Profit per day and week
        per_day = np.round(profit/num_of_days, 2)
        per_week = np.round(profit/num_of_weeks, 2)

        #Filling the list
        start_dates.append(start_date.date())
        end_dates.append(end_date.date())
        profits.append(profit)
        week_profits.append(per_week)
        day_profits.append(per_day)
        days.append(num_of_days)
        weeks.append(num_of_weeks)

        #Creating the DataFrame
    df = pd.DataFrame(data = [start_dates, end_dates, start_prices, last_prices, days, weeks, week_profits, day_profits, profits], index='Start Date, End Date, Bought Price, Sold Price, Days, Approx. Weeks, Weekly, Daily, Total Profit'.split(', ')).transpose()
    return df
```
